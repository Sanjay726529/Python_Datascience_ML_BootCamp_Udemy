---
## 1.4 Training and Testing Data

Now that we've explored the data a bit, let's go ahead and split the data into training and testing sets.


```python
X = c_data.iloc[:, 3:7]
Y = c_data.iloc[:, -1]
```

**Splitting the data into train and test, test size=0.3, random_state=101**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3, random_state=101)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

---
## 1.5 Training the Model


```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression(n_jobs=-1)
lm.fit(x_train, y_train)
```

**coefficients of the model**

```python
temp = pd.DataFrame(lm.coef_, X.columns, columns=['coeff'])
temp
```

## Predicting Test Data


```python
predict = lm.predict(x_test)
```

**scatterplot of the real test values versus the predicted values.**

```python
sns.scatterplot(x=y_test, y=predict)
plt.xlabel('Y_test')
```

---
## 1.6 Evaluating the Model

Let's evaluate our model performance by calculating the residual sum of squares and the explained variance score (R^2).

**Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error**

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

print("MAE: ", mean_absolute_error(y_test, predict))
print("MSE: ", mean_squared_error(y_test, predict))
print("RMSE: ", np.sqrt(mean_squared_error(y_test, predict)))
```

## Residuals distribution



```python
sns.distplot(y_test - predict, bins=30) # almost normally distribted, which means an overall good prediction
```

```python
from sklearn.metrics import r2_score

print("R2_score: ", r2_score(y_test, predict))
```
