```python
import matplotlib.pyplot as plt
import numpy as np
# creates the plots in notebook
%matplotlib inline 
```

**1.1 functional way of using matplotlib**

```python
# simply x,y line plot

x = np.linspace(0,5,10)
y = x**2

plt.plot(x,y)
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Title")
plt.show()
```

```python
# plotting mutiple line graphs within single plot
plt.subplot(1,2,1) # 1 row, 2 column, 1st one
plt.plot(x,y, color='g')
plt.xlabel("X-axis1 ")
plt.ylabel("Y-axis1")
plt.title("Title1")

plt.subplot(1,2,2) # 1 row, 2 column, 2nd one
plt.plot(y,x, color='b')
plt.xlabel("X-axis2")
plt.ylabel("Y-axis2")
plt.title("Title2")

plt.show()
```

**1.2 Object orientied plotting**

```python
fig = plt.figure() # figure object created

# add axes to the figure
ax1 = fig.add_axes([0.1,0.1,1,1])
ax1.plot(x,y)
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.set_title("title")

plt.show()
```

```python
# plot within a plot

fig = plt.figure() # figure object created

# add axes to the figure
ax1 = fig.add_axes([0.1,0.1,1,1])
ax1.plot(x,y)
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.set_title("Big-plot")

# add axes to the figure
ax1 = fig.add_axes([0.2,0.5,0.4,0.4])
ax1.plot(y,x)
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.set_title("small-plot")

plt.show()
```
